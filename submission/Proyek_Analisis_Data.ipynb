````python
import pandas as pd

# Memuat data dari file CSV
day_df = pd.read_csv("day.csv")
hour_df = pd.read_csv("hour.csv")

# Menggabungkan data day.csv dengan hour.csv berdasarkan kolom yang sama
combined_df = pd.merge(
    left=day_df,
    right=hour_df,
    how='inner',
    on=['dteday', 'season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
)

# Menampilkan hasil penggabungan
print(combined_df.head())

combined_df.dropna(axis=0, inplace=True)
combined_df.to_csv("merged_data.csv", index=False)
````

**Insight:**

- Menggabungkan kedua data dari day.csv dan hour.csv menjadi satu dataframe menggunakan function merge()

### MENGIDENTIFIKASI OUTLIERS

Pada grafik boxplot ini membantu kita untuk mengidentifikasi nilai-nilai yang ekstrem atau outliers dalam data suhu, kelembapan, dan jumlah pengendara

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=combined_df[['temp_x', 'hum_x', 'cnt_x']])
plt.title('Identifikasi Outliers Suhu, Kelembapan, dan Jumlah Pengendara')
plt.ylabel('Nilai')
plt.xticks([0, 1, 2], ['Suhu', 'Kelembapan', 'Jumlah Pengendara'])
plt.show()

````

# KORELASI ANTAR VARIABEL

Matriks korelasi menunjukkan seberapa besar hubungan antara variabel suhu, kelembapan, dan jumlah pengendara

````python
correlation_matrix = combined_df[['temp_x', 'hum_x', 'cnt_x']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Variabel')
plt.show()
````

**Insight:**
- Analisis di atas menunjukkan adanya hubungan signifikan antara suhu, kelembapan, dan jumlah pengendara. Rata-rata suhu harian yang lebih tinggi cenderung berhubungan dengan jumlah pengendara yang lebih banyak, sedangkan kelembapan yang tinggi terlihat adanya penurunan jumlah pengendara.


## Visualization & Explanatory Analysis

# Pertanyaan 1: Berapa Rata-rata Suhu Harian Selama Periode Pengamatan?

````python
# Mengubah kolom 'dteday' menjadi datetime dan mengatur sebagai indeks
combined_df['dteday'] = pd.to_datetime(combined_df['dteday'])  # Mengubah ke datetime
combined_df.set_index('dteday', inplace=True)  # Mengatur kolom 'dteday' sebagai indeks

# Menghitung rata-rata harian suhu
daily_temperature = combined_df['temp_x'].resample('D').mean()

# Membuat plot
plt.figure(figsize=(15, 6))
daily_temperature.plot(title='Rata-rata Harian Suhu', xlabel='Tanggal', ylabel='Suhu (°C)')
plt.axhline(y=daily_temperature.mean(), color='r', linestyle='--', label='Rata-rata Suhu Harian')
plt.legend()
plt.show()
````

# Pertanyaan 2: Bagaimana Variasi Jumlah Total (cnt) Berdasarkan Musim?

````python
# Menghitung jumlah total (cnt) berdasarkan musim
seasonal_counts = combined_df.groupby('season')['cnt_x'].sum()

# Membuat plot
plt.figure(figsize=(10, 6))
seasonal_counts.plot(kind='bar', color='skyblue')
plt.title('Jumlah Total Pengendara Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Pengendara')
plt.xticks(ticks=range(4), labels=['Musim Dingin', 'Musim Semi', 'Musim Panas', 'Musim Gugur'], rotation=0)
plt.grid(axis='y')
plt.show()
````

# Pertanyaan 3 : Bagaimana distribusi kelembapan harian?

````python
# Membuat histogram untuk distribusi kelembapan
plt.figure(figsize=(10, 6))
combined_df['hum_x'].hist(bins=30, color='lightblue', edgecolor='black')
plt.title('Distribusi Kelembapan Harian')
plt.xlabel('Kelembapan (%)')
plt.ylabel('Frekuensi')
plt.grid(axis='y')
plt.show()
````

# RFM Analysis

````python
import pandas as pd

# Konversi 'dteday' menjadi datetime
combined_df['dteday'] = pd.to_datetime(combined_df['dteday'])

current_date = combined_df['dteday'].max()
combined_df['recency'] = (current_date - combined_df['dteday']).dt.days
frequency = combined_df.groupby('season')['cnt_x'].count()
monetary = combined_df.groupby('season')['cnt_x'].sum()

rfm_df = pd.DataFrame({
    'Recency': combined_df.groupby('season')['recency'].mean(),
    'Frequency': frequency,
    'Monetary': monetary
})

print(rfm_df)
````
